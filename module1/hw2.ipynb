1. (*6 баллов*) Нужно написать классы ***RussianSentence*** и ***EnglishSentence***.   Конструктор получает на вход предложение на соответствующем языке.   
Атрибуты могут быть любые, подумайте, как вам удобно будет хранить данные о предложении.   
Оба класса должны реализовывать следующие методы :  
    * (*1 балл*) ***get_words()*** - возвращает список слов
    * (*1 балл*) ***get_lemmas()*** - возвращает список лемм
    * (*2 балла*) ***get_pos(pos)*** - принимает на вход название части речи, возвращает все слова этой части речи из предложения в виде списка, в качестве аргумента pos можно передать одну из частей речи из заранее определенного вами набора, одинакового для обоих классов. 
    * (*1 балл*) ***check_affirm()*** - возвращает 'Утвердительное'/'Вопросительное'/'Восклицательное' в зависимости от знака препинания на конце (всё что не вопросительное и не восклицательное давайте считать утвердительным)
    * (*1 балл*) деструктор, который напечатает сообщение о том, что объект удален
2. (*1 балл*) Нужно написать документацию к вашим классам и всем их методам  
3. (*3 балла*) В вашем коде должен быть реализован полиморфизм. Подумайте, где будет логично его использовать. Напишите в комментарии к коду, где и как именно он реализуется.    
      

В случае русского можно токенизировать, лемматизировать и определять части речи с помощью pymorphy, английского - с помощью nltk, но это на ваше усмотрение. На ошибки в определении леммы или части речи можно не обращать внимания.
