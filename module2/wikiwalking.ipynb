```python
from bs4 import BeautifulSoup # pip install bs4
from typing import List
import requests # pip install requests
import urllib

def links_from_text(text):
    soup = BeautifulSoup(text)
    content = soup.find("div", {"id": "mw-content-text"})
    
    links = content.find_all("a")
    
    for link in links:
        href = link.get('href', '')
        
        if href.startswith('/wiki'): 
            yield href # а еще бывают ссылки вида /wiki/какая-то_статья#какой-то_подраздел, мы их считаем за уникальные
            

start_url = u'https://ru.wikipedia.org/wiki/Теория струн'

for url in links_from_text(requests.get(start_url).text):
    print(urllib.parse.unquote(url))
    
```
